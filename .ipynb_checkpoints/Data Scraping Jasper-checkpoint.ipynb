```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
article = 'https://edition.cnn.com/2023/10/23/middleeast/israel-gaza-hamas-war-monday-intl-hnk/index.html'
article1 = 'https://edition.cnn.com/2023/10/17/middleeast/rafah-border-crossing-gaza-israel-explained-intl/index.html'
article3 = 'https://edition.cnn.com/2023/10/21/middleeast/israel-hamas-war-gaza-future-intl/index.html'
article3 = 'https://edition.cnn.com/2023/10/20/middleeast/israel-gaza-hamas-war-friday-intl-hnk/index.html'
article4 = 'https://edition.cnn.com/2023/10/17/middleeast/pregnant-women-suffer-gaza-war-mime-intl/index.html'
article5 = 'https://edition.cnn.com/2023/10/18/middleeast/israel-gaza-hamas-war-wednesday-intl-hnk/index.html'
article6 = 'https://edition.cnn.com/2023/10/16/middleeast/israel-hamas-gaza-war-explained-week-2-mime-intl/index.html'
```

```python
def attributes(article):
    """This function receives an url and returns a DF with different attributes"""
    response = requests.get(article)
    soup = BeautifulSoup(response.content)
    content = soup.find_all("p", attrs = {"class": "paragraph inline-placeholder"})
    # definitions
    text = []
    new_dict= {}
    text_count = {}
    # text and lenght
    for paragraph in content:
        y = paragraph.get_text().strip()
        text.append(y)
    text ="".join(text)
    lenght = len(text)
    # substrings
    new_text = text.lower()
    substrings = ["gaza", "palestine", "hamas", "netanyahu", "terrorism", "migration", "isreal", "attacks", "hostages", "killed"]
    for string in substrings:
        count = new_text.count(string)
        text_count[string] = count
        
    date = soup.find("div", attrs = {"class":"timestamp"}).get_text().strip()
    date = date.split("     ")[-1]
    source = "CNN"
    url = article

    new_dict[0]={"Source":source, "Date" : date, "Text" : text , "Text lenght" : lenght, 
                 "Gaza":text_count["gaza"], "Palestine":text_count["palestine"],"Isreal":text_count["isreal"], 
                 "Hamas":text_count["hamas"], "Netanyahu":text_count["netanyahu"], "Terrorism" : text_count["terrorism"],
                 "Migration":text_count["migration"], "Attacks""URL" : url}
    df = pd.DataFrame.from_dict(new_dict, orient="index")
    return df
```

```python
df_new = pd.concat([attributes(article), attributes(article1), attributes(article2), attributes(article3)], axis=0)
df_new
```
