```python
def attributes(article):
    """This function receives an url and returns a DF with different attributes"""
    response = requests.get(article)
    soup = BeautifulSoup(response.content)
    content = soup.find_all("p", attrs = {"class": "paragraph inline-placeholder"})
    # definitions
    text = []
    new_dict= {}
    text_count = {}
    # text and lenght
    for paragraph in content:
        y = paragraph.get_text().strip()
        text.append(y)
    text ="".join(text)
    lenght = len(text)
    # substrings
    new_text = text.lower()
    substrings = ["gaza", "palestine", "hamas", "netanyahu", "terrorism", "migration", "israel", "attack", "hostage", "killed",
                 "air strike", "biden"]
    for string in substrings:
        count = new_text.count(string)
        text_count[string] = count
    try:   
        date = soup.find("div", attrs = {"class":"timestamp"}).get_text().strip()
        date = date.split("     ")[-1]
    except AttributeError:
        date = None
    
    source = "CNN"
    url = article

    new_dict[0]={"Source":source, "Date" : date, "Text" : text , "Text lenght" : lenght, 
                 "Gaza":text_count["gaza"], "Palestine":text_count["palestine"],"Israel":text_count["israel"], 
                 "Hamas":text_count["hamas"], "Netanyahu":text_count["netanyahu"], "Terrorism" : text_count["terrorism"],
                 "Migration":text_count["migration"], "Attacks": text_count["attack"], "Hostages":text_count["hostage"],
                 "Killed":text_count["killed"],"Air Strike":text_count["air strike"], "Biden": text_count["biden"], "URL" : url}
    df = pd.DataFrame.from_dict(new_dict, orient="index")
    return df
```

```python
dataframes = []
for url in dict_cnn.keys():
    df = attributes(dict_cnn[url])
    dataframes.append(df)

combined_df = pd.concat(dataframes, ignore_index=True)

display(combined_df)
```
